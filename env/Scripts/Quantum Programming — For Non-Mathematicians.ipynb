### Abstract

We deal with different types of probabilities in QC.
- The Marginal Probability is the absolute probability of an event
- The Joint Probability is the probability of two events occurring together
- The Conditional Probability is the probability of one event given the knowledge that another event occurred

In this post, we calculate a **joint probability** with a quantum circuit.

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

def run_circuit(qc, simulator='statevector_simulator', shots=1, hist=True):
    # Tell Qiskit how to simulate our circuit
    backend = Aer.get_backend(simulator)

    # execute the qc
    results = execute(qc,backend, shots=shots).result().get_counts()

    # plot the results
    return plot_histogram(results, figsize=(18,4)) if hist else results
```

```python
# run a test circuit
qc = QuantumCircuit(4)
qc.h([0,1,2,3])
run_circuit(qc)
```

```python
qc = QuantumCircuit(4)
qc.h(0)
run_circuit(qc)
```

## Specify probability using RY gate

```python
from math import asin, sqrt

def prob_to_angle(prob):
    """
    Converts a given P(psi) value into an equivalent theta value.
    """
    return 2*asin(sqrt(prob))
```

```python
# Specify the marginal probability of event A
ev_a = 0.4

qc = QuantumCircuit(4)

# Set qubit to prior
qc.ry(prob_to_angle(ev_a), 0)

run_circuit(qc)
```

## Compute joint probability

```python
# Specify the marginal probabilities
ev_a = 0.4
ev_b = 0.8

qc = QuantumCircuit(4)

# Set event A
qc.ry(prob_to_angle(ev_a), 0)

# Set event B
qc.ry(prob_to_angle(ev_b), 0)

run_circuit(qc)
```

```python
# Specify the marginal probabilities
ev_a = 0.4
ev_b = 0.8

qc = QuantumCircuit(4)

# Set event A
qc.ry(prob_to_angle(ev_a), 0)

# Set event B
qc.ry(prob_to_angle(ev_b), 1)

run_circuit(qc)
```

```python
# Specify the marginal probabilities
ev_a = 0.4
ev_b = 0.8

qc = QuantumCircuit(4)

# Set qubit to prior
qc.ry(prob_to_angle(ev_a), 0)

# Apply half of ev_b
qc.ry(prob_to_angle(ev_b)/2, 1)

# entangle qubits 0 and 1
qc.cx(0,1)

# Apply the other half of ev_b
qc.ry(-prob_to_angle(ev_b)/2, 1)

# unentganle qubits 0 and 1
qc.cx(0,1)

qc.draw()
```

```python
run_circuit(qc)
```

```python
# Specify the marginal probabilities
ev_a = 0.4
ev_b = 0.8

qc = QuantumCircuit(4)

# Set qubit to event A
qc.ry(prob_to_angle(ev_a), 0)

# Apply the controlled RY-gate
qc.cry(prob_to_angle(ev_b), 0, 1)

run_circuit(qc)
```
